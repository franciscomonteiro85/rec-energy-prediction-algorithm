```python
import pandas as pd
import numpy as np
import time
```

```python
weather = pd.read_csv('Weather_Aveiro.csv', sep=";")
weather = weather.rename(columns={'ANO': 'year', 'MS': 'month', 'DI': 'day', 'HR': 'hour', 'MN': 'minute'})
weather['Time'] = pd.to_datetime(weather[['year', 'month', 'day', 'hour', 'minute']])
first_column = weather.pop('Time')
weather.insert(0, 'Time', first_column)
weather.drop(['ESTACAO', 'year', 'month', 'day', 'hour', 'minute'], axis=1, inplace=True)
weather
```

```python
weather_chosen = weather.loc[:,["Time","T_MED","HR_MED","DD_MED","FF_MED","FF_MAX", "T_05_I", "PR_DUR","PR_QTD","PR_I_X","RG_TOT","RD_TOT"]]
weather_chosen = weather_chosen.iloc[576:-4032]
weather_chosen.reset_index(drop=True, inplace=True)
weather_chosen
```

```python
print("NaN values:",weather_chosen.isna().sum())
print("Negative values:\n",(weather_chosen.iloc[:,1:] < 0).sum())
#print("Zero values:\n",(weather_chosen.iloc[:,1:] == 0).sum())
```

```python
# Drop features that have a lot of missing values
weather_final = weather_chosen.drop(["PR_DUR", "RD_TOT"],axis=1)
weather_final
```

```python
def resample_dataset_15_minutes(df):
    df['Time'] = pd.to_datetime(df['Time'])
    start = time.time()
    df_ind = df.set_index('Time')
    df_split = np.array_split(df_ind, len(df_ind)/6)
    df_time = df_ind.resample('15T').apply(lambda x: x[0])
    df_split15 = np.array_split(df_time, len(df_time)/4)
    for hour in range(len(df_split)):
        for col in range(df_time.shape[1]):
            df_split15[hour].iloc[0,col] = df_split[hour].iloc[0,col]
            df_split15[hour].iloc[1,col] = (df_split[hour].iloc[1,col] + df_split[hour].iloc[2,col]) / 2
            df_split15[hour].iloc[2,col] = df_split[hour].iloc[3,col]
            df_split15[hour].iloc[3,col] = (df_split[hour].iloc[4,col] + df_split[hour].iloc[5,col]) / 2
    df_concat = pd.DataFrame(np.concatenate(df_split15))
    dataframe = pd.DataFrame()
    dataframe.index = pd.date_range(start='2022-05-05', end='2023-09-03', closed='left', freq='15T')
    dataframe['Time'] = dataframe.index
    dataframe = dataframe.reset_index(drop=True)
    df_resampled = pd.concat([dataframe, df_concat], axis=1)
    print('Elapsed time: %.4f seconds' % (time.time() - start))
    return df_resampled
```

```python
weather_resample = resample_dataset_15_minutes(weather_final)
weather_resample.columns = weather_final.columns
weather_resample
```

```python
tmp = weather_resample.iloc[:,1:]
tmp[tmp < 0] = np.nan
weather_resample.iloc[:,1:] = tmp
weather_resample
```

```python
print("NaN values:",weather_resample.isna().sum())
#print("Negative values:\n",(weather_resample.iloc[:,1:] < 0).sum())
print("Zero values:\n",(weather_resample.iloc[:,1:] == 0).sum())
```
