```python
import warnings
warnings.filterwarnings('ignore',category=FutureWarning)
warnings.filterwarnings('ignore',category=DeprecationWarning)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate, KFold
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel("../Dataset.xlsx", sheet_name=['Total Consumers'])
df = data['Total Consumers']
df
```

## Auxiliary Functions

```python
def plot_results(preds: np.array, actuals: np.array, title: str):
    
    plt.scatter(actuals, preds, c='b', label='predicted')
    plt.xlabel('actual')
    plt.ylabel('predicted')
    plt.title(title)
    plt.xlim(0, plt.xlim()[1])
    plt.ylim(0, plt.ylim()[1])
    _ = plt.plot([0, 100], [0, 100], '--r', label='y=x')
    plt.legend()
    plt.show()
    
    
def performance_metrics(preds: np.array, actuals: np.array):

    # calculate performance metrics
    
    mse = mean_squared_error(actuals, preds)
    rmse = np.sqrt(mse)
    #wmape = np.sum(np.abs(preds - actuals)) / np.sum(np.abs(actuals)) * 100
    #mape = np.mean(np.abs((actuals - preds) / actuals)) * 100
    mape = np.round(mean_absolute_error(actuals, preds)/actuals.mean(), 2)
    
    r2 = r2_score(actuals, preds)

    # print performance metrics
    print('RMSE: %.4f' % rmse)
    print('MSE: %.4f' % mse)
    print('MAPE: %.4f' % (mape * 100))
    print('R2: %.4f' % r2)
    return rmse, mse,mape,r2

def build_model(estimator, X_train: np.array, y_train: np.array, X_test: np.array):
    
    model = estimator
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    return model, preds
```

## Sum of all houses consumption

```python
df_total = pd.DataFrame(df.sum(axis=1))
df_total.columns = ['Energy']
df_total
```

## Creating 1 hour before DF

```python
X_hour_total = pd.DataFrame()
for i in range(1, 5):
    X_hour_total[f'Energy_{i}'] = df_total.shift(i)
X_hour_total
```

```python
y_total = df_total.copy()
y_total
```

## Clean Data

```python
X_hour_total.dropna(inplace=True)
X_hour_total.reset_index(drop=True, inplace=True)
X_hour_total
```

```python
y_hour_total = pd.DataFrame(y_total[4:])
y_hour_total.reset_index(drop=True, inplace=True)
y_hour_total.columns = ["Energy"]
y_hour_total
```

## Split into train and test

```python
n_train_samples_total = int(len(X_hour_total) * 0.8)
X_train_total = X_hour_total[:n_train_samples_total]
X_test_total = X_hour_total[n_train_samples_total:]
y_train_total = y_hour_total[:n_train_samples_total]
y_test_total = y_hour_total[n_train_samples_total:]
print(X_train_total.shape, X_test_total.shape, y_train_total.shape, y_test_total.shape)
```

```python
fig, axs = plt.subplots(2,2,figsize=(18, 18))
axs[0, 0].set_title('Relationship between last 15 minutes and Y')
axs[0, 0].scatter(X_train_total.values[:,0], y_train_total.values,  color='black')

axs[0, 1].set_title('Relationship between last 30 minutes and Y')
axs[0, 1].scatter(X_train_total.values[:,1], y_train_total.values,  color='red')

axs[1, 0].set_title('Relationship between last 45 minutes and Y')
axs[1, 0].scatter(X_train_total.values[:,2], y_train_total.values,  color='green')

axs[1, 1].set_title('Relationship between last hour and Y')
axs[1, 1].scatter(X_train_total.values[:,3], y_train_total.values,  color='blue')
```

## Build model and predict

```python
model, preds = build_model(LinearRegression(), X_train_total, y_train_total, X_test_total)
rmse, mse, mape, r2 = performance_metrics(preds, y_test_total.values.reshape(-1))
```

```python
plt.figure(figsize=(12, 8))

plot_results(preds, y_test_total, 'Linear Regression model over the last hour')
```
